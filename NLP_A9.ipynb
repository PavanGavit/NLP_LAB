```python
from transformers import pipeline

# FIXED: Use the correct model name for Marathi
# "marathi-sentiment-md" is the Multi-Domain model (Movies, Tweets, etc.)
model_path = "l3cube-pune/marathi-sentiment-md"

print(f"Loading model: {model_path}...")
sentiment_pipeline = pipeline("sentiment-analysis", model=model_path)
print("Model loaded successfully!")

# --- Test with Marathi Sentences ---
data = [
    "ही फिल्म खूप छान आहे.",            # (This film is very good) -> Positive
    "जेवण एकदम बकवास होते.",            # (The food was absolutely rubbish) -> Negative
    "आज हवामान सामान्य आहे.",           # (Today's weather is normal) -> Neutral
    "मला माहित नाही मी काय करावे."       # (I don't know what to do) -> Negative/Neutral
]

print(f"{'MARATHI TEXT':<40} | {'SENTIMENT':<10} | {'SCORE'}")
print("-" * 75)

for sentence in data:
    result = sentiment_pipeline(sentence)[0]

    # Note: L3Cube models usually output "Positive", "Negative", "Neutral"
    # or sometimes numeric labels (0, 1, 2).
    # If you see "LABEL_0", that usually maps to Negative.
    print(f"{sentence:<40} | {result['label']:<10} | {result['score']:.2f}")
```
