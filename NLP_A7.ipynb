```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.probability import FreqDist
from nltk import pos_tag
import string

# Download necessary NLTK resources
# FIXED: Added 'averaged_perceptron_tagger_eng' which is required for newer NLTK versions
resources = [
    "punkt",
    "stopwords",
    "averaged_perceptron_tagger",
    "averaged_perceptron_tagger_eng",  # <--- The missing resource
    "wordnet",
    "omw-1.4",
    "punkt_tab"
]

for res in resources:
    try:
        nltk.download(res, quiet=True)
    except:
        pass

print("NLTK resources downloaded successfully.")
```

```python
class TextAnalyzer:
    def __init__(self):
        self.stop_words = set(stopwords.words('english'))
        self.lemmatizer = WordNetLemmatizer()

    def process(self, text):
        """
        Runs the full NLP pipeline on the input text.
        """
        # 1. SENTENCE TOKENIZATION
        # Breaks text into individual sentences
        sentences = sent_tokenize(text)

        # 2. WORD TOKENIZATION
        # Breaks text into individual words
        words = word_tokenize(text)

        # 3. CLEANING & STOPWORD REMOVAL
        # Remove punctuation and common words (is, the, at) that don't add meaning
        cleaned_words = [
            word for word in words
            if word.lower() not in self.stop_words and word not in string.punctuation
        ]

        # 4. LEMMATIZATION
        # Reduces words to their base form (e.g., "running" -> "run", "better" -> "good")
        lemmatized_words = [self.lemmatizer.lemmatize(word) for word in cleaned_words]

        # 5. POS TAGGING (Part-of-Speech)
        # Identifies if a word is a Noun (NN), Verb (VB), Adjective (JJ), etc.
        # Note: We tag the original words to keep context, not the cleaned list
        pos_tags = pos_tag(words)

        # 6. BASIC ANALYSIS (Frequency Distribution)
        # Counts how often each word appears
        freq_dist = FreqDist(cleaned_words)

        return {
            "sentences": sentences,
            "tokens": words,
            "filtered_tokens": cleaned_words,
            "lemmas": lemmatized_words,
            "pos_tags": pos_tags,
            "top_words": freq_dist.most_common(5)
        }

# Initialize the application
app = TextAnalyzer()
```

```python
# Sample text with diverse punctuation and grammar
sample_text = """
NLTK is a leading platform for building Python programs to work with human language data.
It provides easy-to-use interfaces to over 50 corpora and lexical resources.
Are you learning NLP? Processing text is fun and analyzing data is crucial!
"""

# Run processing
results = app.process(sample_text)

# --- DISPLAY RESULTS ---
print("--- 1. Sentence Tokenization ---")
for i, sent in enumerate(results["sentences"]):
    print(f"Sentence {i+1}: {sent}")

print("\n--- 2. Word Tokenization (First 10) ---")
print(results["tokens"][:10])

print("\n--- 3. Stopword Removal & Cleaning ---")
print(f"Original Count: {len(results['tokens'])} -> Cleaned Count: {len(results['filtered_tokens'])}")
print(results["filtered_tokens"])

print("\n--- 4. POS Tagging (Grammatical Analysis) ---")
# Example Output: ('NLTK', 'NNP') -> NNP means Proper Noun
print(results["pos_tags"][:10])

print("\n--- 5. Top 5 Keywords ---")
for word, count in results["top_words"]:
    print(f"{word}: {count}")
```
