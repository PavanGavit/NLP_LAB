```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# 1. Define the model checkpoint
# This is a specific model trained for English (en) to Hindi (hi)
# For other languages, simply change this string (see list at the bottom of the code)
model_checkpoint = "Helsinki-NLP/opus-mt-en-hi"

# 2. Load Tokenizer and Model
print("Loading model... (This may take a minute)")
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)

print("Model loaded successfully!")
```

```python
def translate_text(text):
    """
    Translates English text to the target Indian language.
    """
    # 1. Tokenize the input text
    # return_tensors="pt" tells it to return PyTorch tensors
    inputs = tokenizer(text, return_tensors="pt", padding=True)

    # 2. Generate Translation
    # This runs the model to predict the corresponding tokens in the target language
    translated_tokens = model.generate(**inputs)

    # 3. Decode the tokens back into text
    translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)

    return translated_text[0]
```

```python
# Sample "Public Information" sentences
sample_texts = [
    "The government has announced a new holiday for all schools tomorrow.",
    "Heavy rainfall is expected in the coastal regions for the next 48 hours.",
    "Please wear a mask and maintain social distancing in public places.",
    "The deadline for submitting tax returns has been extended to next month."
]

print(f"{'ENGLISH (Input)':<70} | {'HINDI (Translated)':<70}")
print("-" * 130)

for text in sample_texts:
    translation = translate_text(text)
    print(f"{text:<70} | {translation:<70}")
```
