```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

# Download necessary NLTK resources
resources = ["wordnet", "omw-1.4", "punkt", "punkt_tab"]
for res in resources:
    try:
        nltk.download(res, quiet=True)
    except:
        pass

print("Resources downloaded.")
```

```python
def get_correct_meaning(sentence, ambiguous_word):
    """
    Uses the Lesk algorithm to find the correct definition of an
    ambiguous word based on the sentence context.
    """
    # 1. Tokenize the sentence (Lesk requires a list of words)
    tokens = word_tokenize(sentence)

    # 2. Apply Lesk Algorithm
    # args: (context_sentence, target_word)
    # returns: A WordNet Synset (the specific meaning)
    best_sense = lesk(tokens, ambiguous_word)

    if best_sense:
        return {
            "word": ambiguous_word,
            "sentence": sentence,
            "sense_name": best_sense.name(),
            "definition": best_sense.definition()
        }
    else:
        return None
```

```python
# List of test cases: (Sentence, The ambiguous word to analyze)
test_cases = [
    # Context 1: Financial Bank
    ("I went to the bank to deposit my money.", "bank"),

    # Context 2: River Bank
    ("The river bank was full of wild flowers.", "bank"),

    # Context 3: Bat (The Animal)
    ("The bat flew out of the cave at night.", "bat"),

    # Context 4: Bat (Sports)
    ("The cricket player broke his bat during the match.", "bat")
]

print(f"{'WORD':<10} | {'CONTEXT & DETECTED MEANING'}")
print("-" * 80)

for sentence, target_word in test_cases:
    result = get_correct_meaning(sentence, target_word)

    if result:
        print(f"Word:      {result['word'].upper()}")
        print(f"Context:   '{result['sentence']}'")
        print(f"Meaning:   {result['definition']}")
        print("-" * 80)
    else:
        print(f"Could not disambiguate '{target_word}' in this context.")
```
